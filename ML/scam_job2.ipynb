```
df['department'] = df['department'].fillna('Missing')
```

```
df['required_experience'] = df['required_experience'].fillna('Missing')
```

```
df['required_education'] = df['required_education'].fillna('Missing')
```

```
freq = df['industry'].value_counts()
rare = freq[freq < 10].index   # anything appearing <10 times

df['industry'] = df['industry'].replace(rare, 'Other')
```

```
df['industry'] = df['industry'].fillna('Missing')
```

```
# 1. Fill missing titles
df['title'] = df['title'].fillna('')

df['title'] = df['title'].str.lower()
df['title'] = df['title'].str.replace('[^a-zA-Z ]', '', regex=True)
df['title'] = df['title'].str.replace(r'\s+', ' ', regex=True)


# 3. Convert titles into numeric form (TF-IDF)
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    stop_words='english',
    max_features=3000
)

X_title = tfidf.fit_transform(df['title'])
```

```
X_title
```

```
# 1. Flag missing location
df['location_missing'] = df['location'].isnull().astype(int)

# 2. Flag incomplete / suspicious location (like "US, ,")
df['location_incomplete'] = df['location'].str.contains('US, ,', na=False).astype(int)

# 3. Remove the original location text
df.drop(columns=['location'], inplace=True)
```

```
df['salary_missing'] = df['salary_range'].isnull().astype(int)
df['salary_suspicious'] = df['salary_range'].str.contains('0-0|500000|1000000', na=False).astype(int)

```

```
df.drop(columns=['salary_range'], inplace=True)
```

```
company_counts = df['company_profile'].value_counts()
```

```
df['company_post_count'] = df['company_profile'].map(company_counts)
df['low_company_activity'] = (df['company_post_count'] <= 2).astype(int)

```

```
df.drop(columns=['company_profile'], inplace=True)
```

```
df['company_post_count'] = df['company_post_count'].fillna(0)
```

```
desc_counts = df['description'].value_counts()
df['description_post_count'] = df['description'].map(desc_counts)
df['reused_description'] = (df['description_post_count'] > 3).astype(int)
```

```
df.drop(columns=['description'], inplace=True)
```

```
benefit_counts = df['benefits'].value_counts()
df['benefits_post_count'] = df['benefits'].map(benefit_counts)
df['reused_benefits'] = (df['benefits_post_count'] > 3).astype(int)
df.drop(columns=['benefits'], inplace=True)
```

```
df['benefits_post_count'] = df['benefits_post_count'].fillna(0)
```

```
req_counts = df['requirements'].value_counts()
df['requirements_post_count'] = df['requirements'].map(req_counts)
df['requirements_post_count'] = df['requirements_post_count'].fillna(0)
df['reused_requirements'] = (df['requirements_post_count'] > 3).astype(int)
df.drop(columns=['requirements'], inplace=True)
```

```
df['description_post_count'] = df['description_post_count'].fillna(0)
```

```
cat_cols = [
    'employment_type',
    'required_experience',
    'required_education',
    'department',
    'industry',
    'function'
]
df = pd.get_dummies(df, columns=cat_cols, drop_first=True)
```

```
df.dtypes
```

```
X_num = df.drop(columns=['fraudulent','title'])

y = df['fraudulent']
```

```
X_num = X_num.select_dtypes(include=['number'])
```

```
bad_cols = X_num.columns[X_num.dtypes == 'object']
bad_cols
```

```
# Convert everything to float
X_num = X_num.astype(float)
```

```
from scipy.sparse import csr_matrix
X_num_sparse = csr_matrix(X_num.to_numpy())
```

```
from scipy.sparse import hstack
x = hstack([X_title, X_num_sparse])
```

```
x
y
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

```
y_pred = model.predict(x_test)
```

```
from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test, y_pred))

```

```
print(classification_report(y_test, y_pred))
```
